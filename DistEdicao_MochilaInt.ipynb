### Distância de edição

Considere as seguintes operações:
- **Substituição:** Trocar uma letra por outra.

Por exemplo, podemos transformar "CALMA" em "PALMA" substituindo o C por P.

- **Remoção:** Remover uma letra.

Por exemplo, podemos transformar "CALMA" em "ALMA" removendo o C.

- **Inserção:** Inserir uma letra.

Por exemplo, podemos transformar "PALMA" em "PALMAS" inserindo um S.

**DEFINIÇÃO**: A **distância de edição** entre uma palavra $x$ e uma palavra $y$ é o menor número de operações de substituição, remoção e inserção necessárias para transformar $x$ em $y$.

Por exemplo, a distância de edição de CACHORRO para CHURRASCO é $6$.

- CACHORRO
- ACHORRO -> remove C
- CHORRO -> remove A
- CHURRO -> substitui O por U
- CHURRAO -> insere A
- CHURRASO -> insere S
- CHURRASCO -> insere C

## Solução com programação dinâmica

Considere que as palavras recebidas são $x_1x_2x_m$ e $y_1y_2...y_n$

### Subproblemas

#### Subproblema E(i,j): calcular custo do melhor alinhamento de $x[1..i]$ com $y[1...j]$

Para $i=0$, consideramos $x[1...0]$ como a palavra vazia. Ou seja,
- $E(0, j) = j$ -- custo da inserção de $j$ caracteres
- $E(i,0) = i$ -- custo da inserção de $i$ caracteres

#### Resposta final: E(m, n) 
#### Como calcular $E(i,j)$?

para as últimas letras $x_i$ e $y_j$ temos 3 possibilidades:

1) $x_i$ é alinhado com $-$: o custo é $1+E(i-1,j)$

2) $y_j$ é alinhado com $-$: o custo é $1+E(i,j-1)$

3) $x_i$ é alinhado com $y_j$: o custo é $E(i-1,j-1)$ mais 1 no caso em que $x_i \neq y_j$

Assim, $E(i,j) = \min(1+E(i-1,j), 1+E(i,j-1), E(i-1,j-1) + diff(i,j))$, onde $diff(i,j)=1$ se $x_i\neq y_j$ e $0$, caso contrário.

Note que $E(i,j)$ depende apenas dos valores $E(i-1,j), 1+E(i,j-1), E(i-1,j-1)$.

Assim, temos o seguinte pseudocódigo:

```
para i de 0 a m faça E(i,0) = i
para j de 0 a n faça E(0,j) = j
para i de 1 a m faça
    para j de 1 a n faça
        E(i,j) = min(1+E(i-1,j), 1+E(i,j-1), E(i-1,j-1) + diff(i,j))

```

````python
def preenche(x,y):
    m = len(x)
    n = len(y)
    #Cria matriz (m+1)x(n+1) inicializada com 0's
    E = [0]*(m+1)*(n+1) 
    E = reshape(E, (m+1,n+1)) 
    
    for i in range(m+1):
        E[i][0] = i
        
    for j in  range(n+1):
        E[0][j] = j
        
    for i in range(1,m+1):
        for j in  range(1,n+1):
            d = 0
            if x[i-1]!=y[j-1]:
                d = 1
            E[i][j] = min(1+ E[i-1][j], 1+ E[i][j-1], E[i-1][j-1]+d)
    print(E)
    print('Resposta: '+str(E[m][n]))
````

````python
preenche("cachorro", "churrasco")
````

## Recursão memoizada

````python
def preenche2(x,y):
    m = len(x)
    n = len(y)
    #Cria matriz (m+1)x(n+1) inicializada com 0's
    E = [0]*(m+1)*(n+1) 
    E = reshape(E, (m+1,n+1)) 
    preencheRec(E, m, n,x,y)
    print(E)
    print(E[m][n])
    
def preencheRec(E, i, j,x,y):
    if E[i][j]!=0:
        return E[i][j]
    if i==0:
        E[i][j]=j
    elif j==0:
        E[i][j]=i
    else:
        d = 0
        if x[i-1]!=y[j-1]:
            d = 1
        E[i][j] = min(1+ preencheRec(E,i-1,j,x,y), 1+ preencheRec(E,i,j-1,x,y), preencheRec(E,i-1,j-1,x,y)+d)
    return E[i][j]
````

````python
preenche2("cachorro","churrasco")
````

## Economizando espaço

Ideia: só precisamos de 2 linhas, pois cada $E(i,j)$ depende apenas da linha atual e da linha anterior!

````python
def preenche3(x,y):
    m = len(x)
    n = len(y)
    #Cria matriz (m+1)x(n+1) inicializada com 0's
    E = [0]*(n+1)*2
    E = reshape(E, (2,n+1)) 
    
    for j in range(n+1):
        E[0][j] = j
        
    for i in range(1,m+1):
        atual=0
        ant = 1
        if i%2==1: 
            atual=1
            ant = 0
        E[atual][0]=i
        for j in  range(1,n+1):
            d = 0
            if x[i-1]!=y[j-1]:
                d = 1
            E[atual][j] = min(1+ E[ant][j], 1+ E[atual][j-1], E[ant][j-1]+d)
    print(E)
    print('Resposta: '+str(E[atual][n]))
````

````python
preenche3("cachorro","churrasco")
preenche("cachorro","churrasco")
````

# Mochila inteira

#### Problema (INT-MOCHILA):
#### Entrada: um inteiro $n$, um inteiro $W$ (capacidade), pesos inteiros $(w_1,\dotsc, w_n)$ e valores $(v_1,\dotsc, v_n)$
#### Soluções viáveis: $(x_1,\dotsc, x_n)$, onde $x_i=0$ ou $x_i=1$ para cada $i$ e $\sum_{i=1}^n x_iw_i \leq W$
#### Objetivo: maximizar $\sum_{i=1}^n x_iv_i$

Já vimos a versão fracionária (em que $0\leq x_i\leq 1$). Como já discutimos, a estratégia de considerar os itens de acordo com custo benefício não funciona.

> **Exercício:** Mostrar um exemplo em que a estratégia gulosa que vimos para a mochila fracionária não devolve o melhor valor para a mochila inteira.

## Subproblema K(W', j): o maior valor para uma mochila de capacidade $W'$ e itens $1,\dotsc, j$

- $K(0,j) = 0$ para todo $j$
- $K(W', 0) = 0$ para todo $W'$
- para $W'>0$ e $j>0$, para $K(W',j)$ temos duas opções:

1) o item $j$ é incluso: valor $v_j + K(W'-w_j, j-1)$ (somente se $w_j\leq W'$)

2) o item $j$ **não** é incluso: valor $K(W', j-1)$

Assim, 
$$
K(W',j) =
\begin{cases}
0,&\text{ se }j=0\text{ ou }W'=0\\
K(W',j-1),&\text{ se }w_j>W'\\
\max\{v_j+ K(W'-w_j,j-1), K(W',j-1)\},&\text{ caso contrário}
\end{cases}
$$


Assim, temos o seguinte pseudocódigo:
```
para W' de 0 a W faça K[W'][0]=0
para j de 0 a n faça K[0][j]=0

para W' de 1 a W
    para j de 1 a n
       se w_j > W'
          K[W'][j]=K[W'][j-1]
       senão
          K[W'][j] = max(v_j+K[W'-w_j][j-1], K[W'][j-1])
```


````python
def mochila(W,w,v):
    n = len(w)
    K = [0]*(W+1)*(n+1)
    K = reshape(K,(W+1,n+1))
    for Wp in range(1,W+1):
        for j in range (1,n+1):
            if w[j-1]>Wp: 
                K[Wp][j]=K[Wp][j-1]
            else:
                K[Wp][j] = max(v[j-1]+K[Wp-w[j-1]][j-1], K[Wp][j-1])
    print(K)
````

````python
mochila(10,[2,3,2,4],[4,4,3,4])
````
